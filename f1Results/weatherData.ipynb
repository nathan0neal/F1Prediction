## Scraping des données méteo

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen as uReq
import re
```

```python
raceURL = {'url': []}

for year in list(range(1950,2021)):

    url = 'https://ergast.com/api/f1/{}.json'
    r = requests.get(url.format(year))
    json = r.json()

    for item in json['MRData']['RaceTable']['Races']:
        try:
            raceURL['url'].append(item['url'])
        except:
            raceURL['url'].append(None)

races = pd.DataFrame(raceURL)
```

```python
races
```

```python
# get the response in the form of html
wikiurl="https://en.wikipedia.org/wiki/2021_Styrian_Grand_Prix"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
info = []
for i,link in enumerate(races.url):
    try:
        print(i)
        uClient = uReq(link)
        page_html = uClient.read()
        uClient.close()
        page_soup = BeautifulSoup(page_html, "lxml")
        my_table = page_soup.find('table',{'class':"infobox vevent"})
        item=my_table.find('th',text=re.compile('Weather')).find_next('td').text
        info.append(item)
    except:
        info.append('NaN')
```

```python
info
```

```python
Weather2 = []
for x in info:
    Weather2.append(x.replace("\n", ""))
Weather = []
for x in Weather2:
    Weather.append(x.replace(".", ""))
Weather
weather = [x.lower() for x in Weather]
weather
```

```python
weatherDict = {'goodWeather': ['dry', 'clear', 'warm', 'hot', 'sunny', 'fine', 'mild'],
                'cold': ['cold', 'fresh', 'chilly', 'cool'],
                'dry': ['dry'],
                'windy': ['windy'],
                'wet': ['showers', 'wet', 'rain', 'drying', 'damp', 'thunderstorms', 'rainy'],
                'cloudy': ['overcast', 'clouds', 'cloudy', 'grey', 'partly']}
```

```python
dfWeather = pd.DataFrame({'Weather': weather})
weather_df = pd.DataFrame(columns = weatherDict.keys())
```

```python
#creating dummy variables
for col in weatherDict:
    weather_df[col] = dfWeather['Weather'].map(lambda x: 1 if any(i in weatherDict[col] for i in x.split()) else 0)
```

```python
weather_df["Race"] = races['url'].apply(lambda x: x.split("wiki/",1)[1])
```

```python
weather_df['year']=weather_df['Race'].apply(lambda x: x.split("_")[0])
```

```python
weather_df.loc[weather_df['year'] == '70th', 'year'] = '2020'
```

```python
weather_df.year = pd.to_numeric(weather_df.year)
```

```python
weather_df.to_json("raceWeather.json")
```
